```python
## https://www.amazon.com/
import re
import time
from selenium import webdriver
from bs4 import BeautifulSoup
driver = webdriver.Chrome("/Users/oiojin831/Dev/likelion/selenium-91/chromedriver_91")
```

```python
url = "https://www.amazon.com/AmazonBasics-Type-C-USB-Male-Cable/dp/B01GGKYO1I"
driver.get(url)
```

```python
all_review_page = driver.find_element_by_xpath('//*[@id="cr-pagination-footer-0"]/a')
all_review_page.click()
```

```python
page = driver.page_source
soup = BeautifulSoup(page, "html.parser")
soup.title
```

```python
all_reviews = soup.find_all("div", class_="a-section celwidget")

all_review_data = []

for page in range(3):
    for review in all_reviews:
        date_list = review.find("span", class_="review-date").text.split(" ")
        day = date_list[-2].replace(',','')
        vote = review.find(attrs={"data-hook":"helpful-vote-statement"})
        review_data = {
            "review": review.find("span", class_="a-size-base review-text review-text-content").text.strip(),
            "name":  review.find("span", class_="a-profile-name").text,
            "rate": review.find("span", class_="a-icon-alt").text.split(" ")[0],
            "date": f'{date_list[-1]}-{date_list[-3]}-{day}',
            "region": re.search("the (.*?) on", review.find("span", class_="review-date").text).group(1),
            "vote": vote.text.split(" ")[0] if vote is not None else "0"
        }
        all_review_data.append(review_data)
    sel_next = driver.find_element_by_xpath('//*[@id="cm_cr-pagination_bar"]/ul/li[2]/a')
    sel_next.click()
    time.sleep(3)

all_review_data
```
